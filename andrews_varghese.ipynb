You have a matrix $X$ of size (10,5) filled with random values. Print the following :
* Minimum value and index where it occured, for each column
* Maximum value and the index where it occured, for each column
* Minimum value and index where it occured, for each row
* Maximum value and index where it occured, for each row
* Minimum value and index where it occured, for the entire matrix
* Maximum value and index where it occured, for the entire matrix

````python
np.random.seed(0)
X = np.random.rand(10,5)

print("\nCOLUMN WISE")
print("MIN")
min_val_col = X.min(0)
min_ind_col = X.argmin(0)
col = 0
for ind in min_ind_col:
    print("Column ", col, " min ", min_val_col[col], " at index ", ind)
    col += 1

print("MAX")
max_val_col = X.max(0)
max_ind_col = X.argmax(0)
col = 0
for ind in max_ind_col:
    print("Column ", col, " max ", max_val_col[col], " at index ", ind)
    col += 1

print("\nROW WISE")
print("MIN")
min_val_row = X.min(1)
min_ind_row = X.argmin(1)
row = 0
for ind in min_ind_row:
    print("Row ", row, " min ", min_val_row[row], " at index ", ind)
    row += 1

print("MAX")
max_val_row = X.max(1)
max_ind_row = X.argmax(1)
row = 0
for ind in max_ind_row:
    print("Row ", row , " max ", max_val_row[row], " at index ", ind)
    row += 1

print("\nOVERALL")
print("Min ", min_val_col.min(), " occurred at ", np.unravel_index(X.argmin(), X.shape))
print("Max ", max_val_col.max(), " occurred at ", np.unravel_index(X.argmax(), X.shape))
````

Let $Y$ denote a square matrix of size 10 with random values. Print the following:
* Determinant
* Rank of the matrix
* Diagonal Elements
* Trace
* Eigen values
* Matrices obtained after Singular Value Decomposition
* Inverse if exists, otherwise Pseudo-Inverse
* Multiplication $Y$ with $X$
* Element-wise multiplication of $Y$ and its inverse(or pseudo-inverse)

````python
np.random.seed(3)
Y = np.random.rand(10,10)

det = np.linalg.det(Y)
print("Det(Y) =", det)
print("\nRank(Y) =", np.linalg.matrix_rank(Y))
print("\nDiagonal(Y) =", Y.diagonal())
print("\nTrace(Y) =", np.trace(Y))
print("\nEigen values(Y) =", np.linalg.eigvals(Y))

u, s, vh = np.linalg.svd(Y)
print("\nSVD \nu:\n", u, "\n\ns:\n", s, "\n\nvh:\n", vh)

if det != 0:
    req_inv = np.linalg.inv(Y)
    print("Inverse Exists and is =\n", req_inv)
else:
    req_inv = np.linalg.pinv(Y)
    print("Inverse DOES NOT Exist. Pseudo Inv is =\n", req_inv)
    
print("Y*X =", np.matmul(Y, X))

print("Element Wise Multiplication =\n", np.multiply(Y, req_inv))
````

In this part, you will be working with the library ```pandas```. Import this library with an alias name ```pd```.

````python
# Initial imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv
Read the data and display the first 5 rows

````python
#Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv

file_url = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv"
df = pd.read_csv(file_url)

print(df.head(5))
````

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````python
need_dummy_for = ['sex', 'embarked', 'pclass']
titanic_data_new = pd.get_dummies(df, columns=need_dummy_for)
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````python
age_mean = titanic_data_new.age.mean()
titanic_data_new['age'].fillna(age_mean, inplace = True)
fare_mean = titanic_data_new.fare.mean()
titanic_data_new['fare'].fillna(fare_mean, inplace = True)
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````python
df_toscale = titanic_data_new.select_dtypes(include='number')
df_scaled = pd.DataFrame(index=df_toscale.index, columns=df_toscale.columns)
for col in df_toscale.columns:
  mean = df_toscale[col].mean()
  std = df_toscale[col].std()
  if std != 0:
    df_scaled[col] = (df_toscale[col] - mean)/std
  else:
    df_scaled[col] = (df_toscale[col] - mean)
````



---
2. Drop the original age, fare columns
3. Append the normalized columns to the titanic_data_new DataFrame.



---



````python
titanic_data_new = titanic_data_new.drop(['age', 'fare'], axis=1)
titanic_data_new = titanic_data_new.assign(**{c: df_scaled[c] for c in df_scaled.columns})
````

###Plotting
1. plot the age against the ID(continuous)
2. plot the fare against the ID(scatter)

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

fig1 = plt.figure()
ax1 = plt.axes()
ax1.set_xlabel('ID')
ax1.set_ylabel('Age')
ax1.set_title('Age vs ID')
ax1.plot(titanic_data_new['age'])

fig2 = plt.figure()
ax2 = plt.axes()
ax2.set_xlabel('ID')
ax2.set_ylabel('Fare')
ax2.set_title('Fare Scatter Plot')
ax2.scatter(x=titanic_data_new.index, y=titanic_data_new['fare'])
````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````python
PI = np.pi
x = np.linspace(-2*PI, 2*PI, num=300)

y1 = np.sin(x)
sp_y1 =  np.fft.fft(y1)
freq_y1 = np.fft.fftfreq(x.shape[-1])

y2 = np.cos(x)
sp_y2 =  np.fft.fft(y2)
freq_y2 = np.fft.fftfreq(x.shape[-1])

fig = plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
ax3 = plt.axes()
ax3.set_xlabel('x')
ax3.set_ylabel('sin(x), cos(x)')
ax3.set_title("Functions")
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()

plt.subplot(2,2,2)
ax4 = plt.axes()
ax4.set_xlabel('f')
ax4.set_ylabel('spectrum sin(x), cos(x)')
ax4.set_title("Spectra")
plt.plot(freq_y1, sp_y1.real, freq_y1, sp_y1.imag)
plt.plot(freq_y2, sp_y2.real, freq_y2, sp_y2.imag)
plt.show()
````
